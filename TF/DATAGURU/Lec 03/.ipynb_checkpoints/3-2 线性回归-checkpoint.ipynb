第一个是3-1中重写写的非线性模型，下面再改变模型，去训练获取线性回归模型。

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.linspace(-0.5, 0.5, 200)[:, np.newaxis]
noise = np.random.normal(0, 0.02, x_data.shape)
y_data =np.square(x_data) + noise
```

```python
plt.figure()
plt.scatter(x_data, y_data)
plt.show()
```

由上图可见，生成的随机数，是成线性的。

```python
x = tf.placeholder(tf.float32, [None, 1])
y = tf.placeholder(tf.float32, [None, 1])
```

```python
Weights_L1 = tf.Variable(tf.random_normal([1, 2]))
Biases_L1 = tf.Variable(tf.zeros([1, 2]))
op_L1 = tf.matmul(x, Weights_L1) + Biases_L1
L1_output = tf.nn.tanh(op_L1)
```

```python
Weights_L2 = tf.Variable(tf.random_normal([2, 1]))
Biases_L2 = tf.Variable(tf.zeros([1,1]))
op_L2 = tf.matmul(L1_output, Weights_L2) + Biases_L2
L2_output = tf.nn.tanh(op_L2)
```

```python
loss = tf.reduce_mean(tf.square(L2_output - y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(0.1)
train = optimizer.minimize(loss)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for epoch in range(200000):
        sess.run(train, feed_dict={x: x_data, y: y_data})
        if epoch % 50000 ==0:
            print(sess.run(loss, feed_dict={x: x_data, y: y_data}))
    
    prediction_value = sess.run(L2_output, feed_dict={x: x_data, y: y_data})
    
    plt.figure()
    plt.scatter(x_data, y_data)
    plt.plot(x_data, prediction_value, 'r-', lw=5)
    plt.show()
```

目标函数改成y=ax+b的线性模型。

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x_data = np.linspace(-0.5, 0.5, 200)[:, np.newaxis]
noise = np.random.normal(0, 0.02, x_data.shape)
y_data =10*x_data + noise


x = tf.placeholder(tf.float32, [None, 1])
y = tf.placeholder(tf.float32, [None, 1])

a = tf.Variable(0.)
b = tf.Variable(0.)

linear_model = a*y + b

loss = tf.reduce_mean(tf.square(linear_model - y))

optimizer = tf.train.GradientDescentOptimizer(0.1)
train = optimizer.minimize(loss)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for epoch in range(200000):
        sess.run(train, feed_dict={x: x_data, y: y_data})
        if epoch % 50000 ==0:
            print(sess.run(loss, feed_dict={x: x_data, y: y_data}))
    
    prediction_value = sess.run(linear_model, feed_dict={x: x_data, y: y_data})
    
    plt.figure()
    plt.scatter(x_data, y_data)
    plt.plot(x_data, prediction_value, 'r-', lw=5)
    plt.show()
```
